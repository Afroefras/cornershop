```python
BASE_DIR = '/Users/efrain.flores/Desktop/EF/Corner/Brands/New_SoM'
FILE_NAME = 'brand_cpg_raw.csv'
```

```python
from pathlib import Path
from pandas import read_csv

BASE_DIR = Path(BASE_DIR)
df = read_csv(BASE_DIR.joinpath(FILE_NAME))
df.columns = list(map(lambda x: x.replace('Product Information ', ''), df.columns))

df.head()
```

```python
df.dropna(subset=['Brand ID'], inplace=True)
df.fillna(value={'CPG Name':'Sin CPG'}, inplace=True)

df.isnull().sum()
```

```python
df['Distinct Product Count'] = df['Distinct Product Count'].astype(str).str.replace(',','').astype(int)
df.sort_values('Distinct Product Count', ascending=False, inplace=True)

df.head()
```

```python
result = df.pivot_table(index=['Brand ID', 'Brand'], aggfunc={'CPG Name':', '.join, 'Distinct Product Count':sum})
result.sample(11)
```

```python
from re import sub

result['CPG'] = result['CPG Name'].map(lambda x: sub(',\sSin\sCPG', '', x))
result['CPG'] = result['CPG'].map(lambda x: sub('^Sin\sCPG,\s', '', x))
result.head()
```

```python
result.to_csv(BASE_DIR.joinpath('brand_cpg.csv'))
```
