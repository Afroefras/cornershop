## Functions

### Timing and tone

```python
import time
from IPython.lib.display import Audio

start = time.time()
def time_exp(x):
    minutes, seconds = np.floor(x/60), 60*(x/60-np.floor(x/60))
    print(f"{'{:.0f}'.format(minutes)} minutos con {'{:.2f}'.format(seconds)} segundos")
    
def tono(a = 1000, b = 700, play_time_seconds = 1, framerate = 4410):
    t = np.linspace(0, play_time_seconds, framerate*play_time_seconds)*np.pi
    return Audio(np.sin(a*t) + np.sin(b*t), rate = framerate, autoplay=True)
```

## Markup

### Split

```python
df['mk_branch_shopper'] = (df['avg_shopper_price'] - df['avg_branch_price']).apply(lambda x:round(x,3))
df['mk_branch_catalog'] = (df['price'] - df['avg_branch_price']).apply(lambda x:round(x,3))
df['mk_branch_supply'] = (df['price_mode'] - df['avg_branch_price']).apply(lambda x:round(x,3))
df['mk_catalog_supply'] = (df['price_mode'] - df['price']).apply(lambda x:round(x,3))
df.fillna({'mk_branch_shopper':0,'mk_branch_catalog':0,
           'mk_branch_supply':0,'mk_catalog_supply':0},inplace=True)
mk = {}

for mk_type in ['mk_branch_shopper','mk_branch_catalog','mk_branch_supply','mk_catalog_supply']:
    mk[mk_type] = df[df[mk_type]!=0].copy()
    print(mk_type,len(mk[mk_type]),'{:.1%}'.format(len(mk[mk_type])/len(df)),
          '\n',mk[mk_type][mk_type].describe(),'\n')
```

### Branch vs shopper

```python
b_s = mk['mk_branch_shopper'].groupby('productbranch_id'
                                     ).agg({'avg_branch_price':'mean','avg_shopper_price':'mean',
                                            'mk_branch_shopper':['mean','sum','count']})
b_s.columns = ['branch_price','shopper_price','markup_un','markup_tot','orders']
b_s.sort_values('markup_tot',inplace=True)
b_s
```

### Branch vs catalog

```python
b_c = mk['mk_branch_catalog'].groupby('productbranch_id'
                                     ).agg({'avg_branch_price':'mean','price':'mean',
                                            'mk_branch_catalog':['mean','sum','count']})
b_c.columns = ['branch_price','catalog_price','markup_un','markup_tot','orders']
b_c.sort_values('markup_tot',inplace=True)
b_c.head(10)
```

### Branch vs supply

```python
b_sc = mk['mk_branch_supply'].groupby('productbranch_id'
                                     ).agg({'avg_branch_price':'mean','price_mode':'mean',
                                            'mk_branch_supply':['mean','sum','count']})
b_sc.columns = ['branch_price','supply_price','markup_un','markup_tot','orders']
b_sc.sort_values('markup_tot',inplace=True)
b_sc.head(22)
```

### Branch vs supply (product_id)

```python
b_c_p = mk['mk_branch_supply'].groupby('catalog_product_id'
                                     ).agg({'avg_branch_price':'mean','price_mode':'mean',
                                            'mk_branch_supply':['mean','sum','count']})
b_c_p.columns = ['branch_price','supply_price','markup_un','markup_tot','orders']
b_c_p.sort_values('markup_tot',inplace=True)
b_c_p.head(22)
```

### Catalog vs supply

```python
b_sc = mk['mk_branch_catalog'].groupby('productbranch_id'
                                     ).agg({'price':'mean','price_mode':'mean',
                                            'mk_catalog_supply':['mean','sum','count']})
b_sc.columns = ['catalog_price','supply_price','markup_un','markup_tot','orders']
b_sc.sort_values('markup_tot',inplace=True)
b_sc.head(10)
```
