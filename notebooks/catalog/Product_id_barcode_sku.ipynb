```python
import os
import pandas as pd

df = pd.read_csv(os.path.join(BASE_DIR,FILE_NAME))
df.sample()
```

```python
import numpy as np

def sku_batches(data,col,n):
    df = data.dropna(subset=[col]).reset_index()
    x = 0
    for i in range(int(np.ceil(len(data)/n))):
        print("'"+"','".join(df.loc[x:x+n,col].astype(str).values)+"'",'\n'*2,'-'*100,'\n')
        x += n
```

```python
sku_batches(df,COL,N)
```
