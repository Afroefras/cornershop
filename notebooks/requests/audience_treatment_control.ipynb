# Par√°metros

```python
from pathlib import Path

BASE_DIR = Path('/Users/efraflores/Desktop/EF/Corner/Requests/audience')
USERS_FILE = 'user_uuids.csv'
ORDERS_FILE = 'apr_orders.csv'
BRANCHES_FILE = 'branches.csv'
```

# Lectura

```python
from pandas import read_csv

users = read_csv(BASE_DIR.joinpath(USERS_FILE))
users.head()
```

```python
orders = read_csv(BASE_DIR.joinpath(ORDERS_FILE))
orders.head()
```

```python
branches = read_csv(BASE_DIR.joinpath(BRANCHES_FILE))
branches.head()
```

# Uni√≥n

```python
from functools import reduce

df = reduce(lambda x,y: x.merge(y), [orders,branches,users])
df.head()
```

# Resumen

```python
from pandas import Series

aux = df[
    # (df['store'] == 'Soriana')&
    (df['order_date'] >= '15/04/22')&
    (df['order_date'] <= '20/04/22')
    ].copy()
aux.groupby(['name', 'store'])[['order_id', 'user_id']].agg(Series.nunique)
```
